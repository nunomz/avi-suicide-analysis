```python
import pandas as pd
from functools import reduce
```

# Main datasets creation
This notebook creates two datasets:
- mortes_causa: contains deaths by cause
- main: contains suicide specific info

### upload all datasets

```python
df_mortes_causa = pd.read_csv("dados/our-world-in-data/annual-number-of-deaths-by-cause.csv") #7273
df_continents = pd.read_csv("dados\our-world-in-data\continents-according-to-our-world-in-data.csv")
```

```python
df_ratio = pd.read_csv("dados\our-world-in-data\share-deaths-suicide.csv")
df_mf_ratio = pd.read_csv("dados\our-world-in-data\Male-Female-Ratio-of-Suicide-Rates.csv") #6188
df_mf = pd.read_csv("dados\our-world-in-data\male-vs-female-suicide.csv")
#df_pesticide nao estamos a usar
df_sex = pd.read_csv("dados\our-world-in-data\suicide-death-rates-by-sex.csv")
df_raties = pd.read_csv("dados\our-world-in-data\suicide-death-rates.csv")
#suicide rate 1990 nao entendo 
df_firearms = pd.read_csv("dados\our-world-in-data\suicide-rate-by-firearm.csv")
df_ages = pd.read_csv("dados\our-world-in-data\suicide-rates-by-age-detailed.csv")
```

### create list with all datasets and merge them

```python
data_frames = [df_ratio, df_mf_ratio, df_mf, df_sex, df_raties, df_firearms, df_ages]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right), data_frames)
df_merged
```

### fix continent info

```python
df_continents = df_continents.drop(columns=['Entity','Year'])
df_continents
```

```python
df_merged = df_merged.drop(columns=['Continent'])
df_main = df_merged.merge(df_continents,how='left',left_on='Code',right_on='Code')
df_main
```

```python
df_mortes_causa = df_mortes_causa.merge(df_continents, how='left', left_on='Code', right_on='Code')
df_mortes_causa
```

### reorganize columns

```python
continent = df_main.pop('Continent')
df_main.insert(3, 'Continent', continent)
#df_main
```

```python
continent2 = df_mortes_causa.pop('Continent')
df_mortes_causa.insert(3, 'Continent', continent2)
#df_mortes_causa
```

### write dataframes to csv

```python
df_mortes_causa.to_csv('mortes_causa.csv')
```

```python
df_main.to_csv('main.csv')
```
